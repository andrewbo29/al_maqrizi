```python
def otsu(img, radius=20):
    selem = skimage.morphology.disk(radius)
    threshold_global_otsu = skimage.filters.threshold_otsu(img)
    global_otsu = img >= threshold_global_otsu
    return global_otsu * 1
```

```python
def accept_region_prop(rp):
    if rp.minor_axis_length == 0:
        return False
    if rp.major_axis_length / rp.minor_axis_length > 10:
        return False
    if rp.major_axis_length < 8:
        return False
    if rp.major_axis_length > 224:
        return False
    return True


def filter_regions_by_properties(regions):
    return filter(accept_region_prop, regions)


from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler


def filter_regions_by_position(regions):
    X = np.array([rp.centroid for rp in regions])
    X = StandardScaler().fit_transform(X)
    db = DBSCAN(eps=0.5, min_samples=10).fit(X)
    labels = db.labels_
    
    if max(labels) == -1:
        return regions
    else:
        return [r for r, l in zip(regions, labels) if l >= 0]
    

norm = matplotlib.colors.Normalize(vmin=-20, vmax=10)
cmap = matplotlib.cm.hot
x = 0.3

m = matplotlib.cm.ScalarMappable(norm=norm, cmap=cmap)
print m.to_rgba(x)


def extract_regions(image):
    label_img = skimage.measure.label(image)
    regions = skimage.measure.regionprops(label_img)
    regions_filtered = filter_regions_by_properties(regions)
    regions_filtered = filter_regions_by_position(regions_filtered)
    return regions_filtered


def plot_image_regions(image, regions=None):
    if regions is None:
        regions = extract_regions(image)

    fig, ax = plt.subplots()
    ax.imshow(1 - image, cmap=plt.cm.gray)
    
    for i, props in enumerate(regions):

        y0, x0 = props.centroid
        orientation = props.orientation
        x1 = x0 + math.cos(orientation) * 0.5 * props.major_axis_length
        y1 = y0 - math.sin(orientation) * 0.5 * props.major_axis_length
        x2 = x0 - math.sin(orientation) * 0.5 * props.minor_axis_length
        y2 = y0 - math.cos(orientation) * 0.5 * props.minor_axis_length

        ax.plot((x0, x1), (y0, y1), '-r', linewidth=.5)
        ax.plot((x0, x2), (y0, y2), '-r', linewidth=.5)
        ax.plot(x0, y0, '.g', markersize=5)

        minr, minc, maxr, maxc = props.bbox
        c = 0.1 * (maxc - minc)
        r = 0.1 * (maxr - minr)
#         minr -= r * 0.1
#         maxr += r * 0.1
#         minc -= c * 0.1
#         maxc += c * 0.1
        bx = (minc, maxc, maxc, minc, minc)
        by = (minr, minr, maxr, maxr, minr)
#         col = i / 20.
        ax.plot(bx, by, '-b', linewidth=1) #, color=m.to_rgba(col))
    
```

```python
def new_image_from_regions(image, regions, grab_extra=1.1):
    image_new = np.zeros(image.shape)
    for r in regions:
        minr, minc, maxr, maxc = r.bbox
        c_w = int((maxc - minc) * grab_extra / 2.)
        r_w = int((maxr - minr) * grab_extra / 2.)
        r, c = r.centroid
        minr = r - r_w
        maxr = r + r_w
        minc = c - c_w
        maxc = c + c_w
        image_new[minr:maxr,minc:maxc] = image[minr:maxr,minc:maxc]
    return image_new


def plot_cleared_image(image):
    new_image = new_image_from_regions(image, extract_regions(image))
    plt.imshow(new_image)
```

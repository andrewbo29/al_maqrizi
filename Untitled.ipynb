```python
def otsu(img, radius=20):
    selem = skimage.morphology.disk(radius)
    threshold_global_otsu = skimage.filters.threshold_otsu(img)
    global_otsu = img >= threshold_global_otsu
    return global_otsu * 1
```

```python
def accept_region_prop(rp):
    if rp.minor_axis_length == 0:
        return False
    if rp.major_axis_length / rp.minor_axis_length > 10:
        return False
    if rp.major_axis_length < 15:
        return False
    if rp.major_axis_length > 160:
        return False
    return True


def plot_image_regions(image):
    label_img = skimage.measure.label(image)
    regions = skimage.measure.regionprops(label_img)

    fig, ax = plt.subplots()
    ax.imshow(image, cmap=plt.cm.gray)

    def accept_region_prop(rp):
        if rp.minor_axis_length == 0:
            return False
        if rp.major_axis_length / rp.minor_axis_length > 10:
            return False
        if rp.major_axis_length < 15:
            return False
        if rp.major_axis_length > 160:
            return False
        return True

    for props in regions:
        if not accept_region_prop(props):
            continue

#         y0, x0 = props.centroid
#         orientation = props.orientation
#         x1 = x0 + math.cos(orientation) * 0.5 * props.major_axis_length
#         y1 = y0 - math.sin(orientation) * 0.5 * props.major_axis_length
#         x2 = x0 - math.sin(orientation) * 0.5 * props.minor_axis_length
#         y2 = y0 - math.cos(orientation) * 0.5 * props.minor_axis_length

#         ax.plot((x0, x1), (y0, y1), '-r', linewidth=2.5)
#         ax.plot((x0, x2), (y0, y2), '-r', linewidth=2.5)
#         ax.plot(x0, y0, '.g', markersize=15)

        minr, minc, maxr, maxc = props.bbox
        bx = (minc, maxc, maxc, minc, minc)
        by = (minr, minr, maxr, maxr, minr)
        ax.plot(bx, by, '-b', linewidth=2.5)
```

```python
fname = 'data/al_maqrizi/Isl. Ms. 605 University of Michigan Library (Special Collections Library)/UM605 pgs. 002-003.JPG'
image = 1 - otsu(skimage.io.imread(fname, as_grey=True))
plot_image_regions(image)
```
